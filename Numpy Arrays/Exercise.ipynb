Valid Sudoku

```python
s = np.array([
    [ 5, 3, 4, 6, 7, 8, 9, 1, 2 ],
    [ 6, 7, 2, 1, 9, 5, 3, 4, 8 ],
    [ 1, 9, 8, 3, 4, 2, 5, 6, 7 ],

    [ 8, 5, 9, 7, 6, 1, 4, 2, 3 ],
    [ 4, 2, 6, 8, 5, 3, 7, 9, 1 ],
    [ 7, 1, 3, 9, 2, 4, 8, 5, 6 ],

    [ 9, 6, 1, 5, 3, 7, 2, 8, 4 ],
    [ 2, 8, 7, 4, 1, 9, 6, 3, 5 ],
    [ 3, 4, 5, 2, 8, 6, 1, 7, 9 ],
])
```

```python
rowSum = np.sum(s, axis=0)
for i in rowSum:
    if i != 45:
        print("Sudoku is not valid row wise") 
        break
else:
    print("Sudoku is valid row wise") 
```

```python
colSum = np.sum(s, axis=1)
for i in colSum:
    if i != 45:
        print("Sudoku is not valid column wise") 
        break
else:
    print("Sudoku is valid column wise") 
```

```python
# s[0:3,0:3].sum()
# s[0:3,0:3].sum()
for i in range(0,9,3):
    for j in range(0,9,3):
        # print(f"[{i}:{i+3}, {j}:{j+3}]")
        if s[i:i+3, j:j+3].sum() != 45:
            print("Each 3 x 3 block sum is not 45, Hence it is not valid sudoku")
            break
else:   
    print("Each 3 x 3 block sum is 45, Hence it is a valid sudoku")

    
```
