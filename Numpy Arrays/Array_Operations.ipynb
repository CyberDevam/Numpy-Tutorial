ADVANCE ARRAY MANIPULATION

Stacking Array

```python
# in stacking array we use vectors only to create matrix or vector
a = np.arange(1,5)
b = np.arange(5,9)
```

```python
a
```

```python
b
```

```python
# 1. vertical stack
np.vstack((a,b)) # vertically dono arrays ko add krenge with no changes in array
```

```python
# 1. horizontal stack
np.hstack((a,b)) # will come in a single vector
```

```python
np.column_stack((a,b)) # same index will in the one row, can say a row with same index of vector input
```
