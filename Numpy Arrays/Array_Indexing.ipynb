NUMPY INDEXING AND SLICING OF MATRIX

```python
arr = np.arange(1,31)
arr = arr.reshape(6,5)
arr
```

```python
# to get row
arr[0] # will give whole first row
arr[3]
```

```python
# will give a specific element
arr[0,2]
```

```python
# slicing
""" 
i need ->
    2, 3
    7, 8
"""
arr[0:2, 1:3]
```

```python
""" 
i need ->
    19, 20
    24, 25
    29, 30
"""
arr[3:,3:]
```

```python
# yess the funniest thing which was not in the game i.e., how to retrieve the cols
arr[:, 2] # : -> to get whole row right!!! nothing but a vector
```

BOOLEAN INDEXING

```python
# small array operation
arr = np.arange(11,21)
arr
```

```python
bool_index = arr % 2 == 0
bool_index # will provide the operation which is performed as in array it is true or not
```

```python
arr = arr[bool_index]
arr
```

```python
arr = np.arange(11,21)
bool_index = arr % 2 != 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
